```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
n_imag = 10
cats_train = []
dogs_train = []
for i in range(n_imag):
    cats_train.append(mpimg.imread('/Users/juda/Documents/Codes/machine_learning/PetImages/Cat/'+str(i)+'.jpg')[...,:3])
    dogs_train.append(mpimg.imread('/Users/juda/Documents/Codes/machine_learning/PetImages/Dog/'+str(i)+'.jpg')[...,:3])
```

```python
cats_lin = []
dogs_lin = []
len_cats = 0
len_dogs = 0
for i in range(n_imag):
    s_c = np.shape(cats_train[i])[0]*np.shape(cats_train[i])[1]*3
    s_d = np.shape(dogs_train[i])[0]*np.shape(dogs_train[i])[1]*3
    cats_lin.append(cats_train[i].reshape(s_c))
    dogs_lin.append(dogs_train[i].reshape(s_d))
    len_cats += s_c
    len_dogs += s_d
```

```python
data = np.hstack((np.vstack((dogs_lin,np.zeros(len(dogs_lin)))),np.vstack((cats_lin,np.ones(len(cats_lin))))))
```
